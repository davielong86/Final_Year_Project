```python


hd_DQT = temp.describe().T.round(2)
hd_DQT.index.name = "Features"

#show the table
hd_DQT.round(3)
```

```python
# Calculating the cardinality for each feature
arr_index = hd_DQT.index
arr_cardinality = []

for row in arr_index:
    val = len(temp[row].unique())
    arr_cardinality.append(val)

hd_DQT['cardinality'] = arr_cardinality
hd_DQT
```

```python
hd_DQT['median'] = temp.median()
hd_DQT
```

```python
hd_DQT = hd_DQT[['count','cardinality','min','25%','mean','median','75%','max','std']]
hd_DQT.columns = ['Count','Cardinality','Min','1st Quart','Mean','Median','3rd Quart','Max','Std. Dev.']
hd_DQT
```
